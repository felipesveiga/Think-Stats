<h2 style='font-size:30px'> An Introduction to Continuous Probability Distributions</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Uma função de distribuição probabilística demonstra a probabilidade de um valor pertencente a uma distribuição se encontrar em um certo intervalo. A curva da função busca demonstrar o mesmo tipo de informação que um histograma.
        </li>
    </ul>
</div>

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(100)
# Observe o gráfico; ele demonstra a probabilidade de uma variável aleatória da distribuição 'x' se encontrar dentro de um intervalo.
# Isso é obtido medindo a integral da função.
sns.kdeplot(x)
```

<h2 style='font-size:30px'> An Introduction to the Continuous Uniform Distribution</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Uma distribuição uniforme é aquela em que a probabilidade para uma variável é igual para todos os intervalos igualmente espaçados. O gráfico se parecerá com um retângulo.
        </li>
    </ul>
</div>

```python
from scipy.stats import uniform
x = np.random.uniform(0,1, size=1000000)
sns.kdeplot(x)
```

<h2 style='font-size:30px'> An Introduction to the Normal Distribution</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Uma distribuição Gaussiana ou Normal é aquela em que seus valores tendem a convergir à sua média.
        </li>
    </ul>
</div>

```python
# A Distribuição Normal Padrão possui uma média 0 e variância 1.
x = np.random.normal(loc=0, scale=1, size=500)
sns.kdeplot(x);
```

<center> 
    <h1> Notação para Variáveis Aleatórias que São Distribuições Normais</h1>
    <img src='normal1.png'>
</center>

```python
s = np.random.normal(loc=2.3, scale=.3,size=10000)
std_s = (s - s.mean()) / s.std()

sns.kdeplot(std_s)
```

<h2 style='font-size:30px'> An Introduction to the Chi-Square Distribution</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A distribuição Chi-Quadrado surge da soma de k distribuições normais padrão ao quadrado. O valor de 'k' denota a quantidade de graus de liberdade que a variável aleatória possuirá.
        </li>
        <li> 
            O valor de 'k' também corresponderá à média da distribuição!
        </li>
    </ul>
</div>

```python
# Montando uma distribuição chi-quadrado com um grau de liberdade (ou seja, formada por uma única distribuição normal padrão ao quadrado).
x_2 = np.random.chisquare(1, size=9999999)

# Medindo a média da distribuição:
print(f'Média: {x_2.mean()}')

# Observe que o formato da distribuição se assemelha a um véu de noiva.
sns.kdeplot(x_2)
```

```python
# Conforme aumentamos o grau de liberdade da distribuição, o formato de véu é perdido.
for df in [1, 3, 10]:
    sns.kdeplot(np.random.chisquare(df, size=1000), label=f'df={df}')

plt.legend()
```

<h2 style='font-size:30px'> An Introduction to the F Distribution</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A distribuição F surge a partir de duas distribuições Chi-Quadrado.
        </li>
        <li> 
            Sendo U1 e U2 distribuições Chi-Quadrado com graus de liberdade $\mu 1$ e $\mu 2$, a distribuição F será $F=\frac{\frac{U1}{\mu 1}}{\frac{U2}{\mu 2}}$
        </li>
    </ul>
</div>

```python
a = np.random.f(50,50,1000)
sns.kdeplot(a)
```
