<h2 style='font-size:30px'> An Introduction to Continuous Probability Distributions</h2>
<div>

        <li> 
            Uma função de distribuição probabilística demonstra a probabilidade de um valor pertencente a uma distribuição se encontrar em um certo intervalo. A curva da função busca demonstrar o mesmo tipo de informação que um histograma.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(100)
# Observe o gráfico; ele demonstra a probabilidade de uma variável aleatória da distribuição 'x' se encontrar dentro de um intervalo.
# Isso é obtido medindo a integral da função.
sns.kdeplot(x)
```

<div>
    <hr>
    <h2 style='font-size:30px'> An Introduction to the Chi-Square Distribution</h2>
</div>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A distribuição Chi-Quadrado surge da soma de k distribuições normais padrão ao quadrado. O valor de 'k' denota a quantidade de graus de liberdade que a variável aleatória possuirá.
        </li>
        <li> 
            O valor de 'k' também corresponderá à média da distribuição!
        </li>
    </ul>
</div>

```python
# Montando uma distribuição chi-quadrado com um grau de liberdade (ou seja, formada por uma única distribuição normal padrão ao quadrado).
x_2 = np.random.chisquare(1, size=9999999)

# Medindo a média da distribuição:
print(f'Média: {x_2.mean()}')

# Observe que o formato da distribuição se assemelha a um véu de noiva.
sns.kdeplot(x_2)
```

```python
# Conforme aumentamos o grau de liberdade da distribuição, o formato de véu é perdido.
for df in [1, 3, 10]:
    sns.kdeplot(np.random.chisquare(df, size=1000), label=f'df={df}')

plt.legend()
```

<div>
    <hr>
    <h2 style='font-size:30px'> An Introduction to the F Distribution</h2>
</div>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A distribuição F surge a partir de duas distribuições Chi-Quadrado.
        </li>
        <li> 
            Sendo U1 e U2 distribuições Chi-Quadrado com graus de liberdade $\mu 1$ e $\mu 2$, a distribuição será $F=\frac{\frac{U1}{\mu 1}}{\frac{U2}{\mu 2}}$ com $(\mu 1, \mu 2)$ graus de liberdade.
        </li>
    </ul>
</div>

```python
# Criando uma distribuição F com (8,3) graus de liberdade.
f = np.random.f(8,3,10000)
sns.kdeplot(f)
plt.xlim([0,30])
```

```python
# Observe, distribuições com graus de liberdade trocados não são idênticas!
sns.kdeplot(np.random.f(8,3, size=1000), label='ddof=(8,3)')
sns.kdeplot(np.random.f(3,8, size=1000), label='ddof=(3,8)')
plt.xlim([0,30])
plt.legend()
```

<div> 
    <ul style='font-size:20px'> 
        <li> 
            Quando $\mu 2 > 2$, a média da distribuição F será de $\mu=\frac{\mu 2}{\mu 2 -2}$
        </li>
    </ul>
</div>

```python
f.mean(), 3/(3-2)
```

<div> 
    <ul style='font-size:20px'> 
        <li> 
            Quando $\mu 1 = \mu 2$, a mediana da distribuição F será 1!
        </li>
    </ul>
</div>

```python
# Ambas as Chi-Quadrado terão 5 graus de liberdade.
f_equal = np.random.f(5,5, size=10000)
np.median(f_equal)
```

<div>

    <h2 style='font-size:30px'> Exponential Distribution</h2>
</div>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A Distribuição Exponencial é considerada o inverso da Poisson. Sua principal aplicação é no descobrimento do tempo necessário para que um evento ocorra. Como o tempo possui uma continuidade, a Distribuição Exponencial é, diferentemente da Poisson, uma distribuição <strong> contínua</strong>. 
        </li>
        <li> 
            Os eventos devem todos ocorrer de maneira independente e a sua probabilidade de ocorrência não pode se alterar.
        </li>
        <li> 
            Segue a PDF da Distribuição Exponencial (não é PMF, pois trata-se de uma distribuição contínua):
            $$
                P(X=x)=\lambda e^{-\lambda x} | \lambda=\frac{1}{\mu}
            $$
        </li>
        <li> 
            Segue a CDF da distribuição:
            $$
                P(X \leq x)=1-e^{-\lambda x}
            $$
        </li>
    </ul>
</div>

```python
def factorial(n):
    if n == 0:
        return 1
    for i in range(1,n):
        n *=i
    return n
```
